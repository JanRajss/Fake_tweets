```python
def compare_algorithms2df(MLA, X_train, X_test, y_train, y_test, sorted_by_measure='accuracy'):
  
    #show grid with compared results - accuracy, recall, ppv, f1-measure, mcc

  MLA_columns = []
  MLA_compare = pd.DataFrame(columns = MLA_columns)


  row_index = 0
  for alg in MLA:
    X_test = pd.DataFrame(X_test)
    y_test = pd.DataFrame(y_test)
    X_train = pd.DataFrame(X_train)
    y_train = pd.DataFrame(y_train)
    predicted = alg.predict(X_test)

    # MLA_name = alg.__class__.__name__

    MLA_name = list(pipe_dict.values())[row_index]
    MLA_compare.loc[row_index,'MLA Name'] = MLA_name
    MLA_compare.loc[row_index, 'MLA Train Accuracy'] = round(alg.score(X_train, y_train), 4)
    MLA_compare.loc[row_index, 'MLA Test Accuracy'] = round(alg.score(X_test, y_test), 4)
    MLA_compare.loc[row_index, 'MLA Precission'] = precision_score(y_test, predicted, average='weighted')
    MLA_compare.loc[row_index, 'MLA Recall'] = recall_score(y_test, predicted, average='weighted')
    MLA_compare.loc[row_index, 'f1_score'] = f1_score(y_test, predicted, average='weighted')
    row_index+=1

  MLA_compare.sort_values(by = [sorted_by_measure], ascending = False, inplace = True)
  return MLA_compare


```

```python
# Comparing algorithms

compare_algorithms2df(MLA, X_train, X_test, y_train, y_test, sorted_by_measure='f1_score')
```
